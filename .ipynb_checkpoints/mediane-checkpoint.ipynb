```python
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import squarify 

%matplotlib inline
```

```python
df = pd.read_csv('lille_med.csv', sep=';').sort_values(by=['mediane'])
```

```python
# create a color palette, mapped to these values
cmap = matplotlib.cm.Greens
mini=min(df['mediane'])
maxi=max(df['mediane'])
norm = matplotlib.colors.Normalize(vmin=mini, vmax=maxi)
colors = [cmap(norm(value)) for value in df['mediane']]
```

```python
# labels for squares
labels = ["%s\n%s â‚¬" % (label) for label in zip(df['Nom'], df['mediane'])]
```

```python
plt.figure(figsize=(20,10))
squarify.plot(sizes=df['mediane'], label=labels, alpha=.8, color=colors  )
plt.title("MÃ©diane des revenus disponibles par UC - 2018",fontsize=23,fontweight="bold")
plt.axis('off')
plt.show()
```
